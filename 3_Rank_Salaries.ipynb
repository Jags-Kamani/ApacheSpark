Question:

Write a solution to rank the salaries from the Employee table according to the following rules:

Salaries should be ranked from the highest to the lowest.

If there is a tie between two salaries, both should have the same ranking.

After a tie, the next ranking number should be the next consecutive integer value (i.e., no holes in the rank sequence).

Return the top 25 ranked salaries, ordered by salary in descending order.

````python
data = {
    'id' : [1, 2, 3, 4, 5],
    'salary' : [7000, 9000, 6000, 9000, 8000]
}
````

Sample Input                    

---


```
| id | salary |
| -- | ------ |
| 1  | 7000   |
| 2  | 9000   |
| 3  | 6000   |
| 4  | 9000   |
| 5  | 8000   |

```




Sample output



```
| salary | rank |
| ------ | ---- |
| 9000   | 1    |
| 9000   | 1    |
| 8000   | 2    |
| 7000   | 3    |
| 6000   | 4    |

```



**Your Solution Here**

````python
import pandas as pd

employee = pd.DataFrame(data)
print(employee)

employee['rank'] = employee['salary'].rank(method = 'dense', ascending = False)
ordered = employee.sort_values(by=['salary', 'rank'], ascending = False)
result = ordered[['salary', 'rank']]
print(result)
````
